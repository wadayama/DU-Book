```julia
using Convex, SCS
items  = [:Gold, :Silver, :Bronze]
values = [5.0, 3.0, 1.0]
weights = [2.0, 1.5, 0.3]

# Define a variable of size 3, each index representing an item
x = Variable(3)
p = maximize(x' * values, 0 <= x, x <= 1, x' * weights <= 3)
solve!(p, SCS.Optimizer)
println([items x.value])
```

```julia
n = 4
x = Variable(n)
cons = [x >= -1, x <= 1]
#cons = [sumsquares(x) <= 1]
for i in 1:100
    Q = randn(n, n)
    Q = Q'*Q
    q = randn(n)
    obj = quadform(x,Q) + dot(x,q)
    problem = minimize(obj, cons)
    solve!(problem, SCS.Optimizer; silent_solver = true)
    println(evaluate(x))
    #println(norm(evaluate(x)))
end
```

```julia
function mini_batch()
    x = Variable(n)
    cons = [x >= -1, x <= 1]
    Q = randn(n, n)
    Q = Q'*Q
    q = randn(n)
    obj = 0.5*quadform(x,Q) + dot(x,q)
    problem = minimize(obj, cons)
    solve!(problem, SCS.Optimizer; silent_solver = true)
    return Q, q, evaluate(x)
end
```

```julia
Q, q, sol = mini_batch()
```

```julia
b(x) = -log(-x+1)-log(x+1)
db(x) = 1/(-x+1) -1/(x+1)

function GD(Q, q)
    x = zeros(n)
    for i in 1:100
        μ = 2.0/i
        x = x - 0.05*(Q*x + q + μ*db.(x))
    end
    return x
end
```

```julia
GD(Q,q)
```
