```julia
using LinearAlgebra
using Plots
gr()
using Random
Random.seed!(1)
using Flux
```

### 行列パラメータの設定

以下の行列パラメータは、文献「深層展開による出力フィードバック安定化の検討」小林他(2019)に基づく。

```julia
A = [-0.2 0.4 -1.0 -0.4;
    -0.7 -0.8 -0.6 -0.3;
    -0.2 0.1 -0.2 0.4;
    -0.6 0.8 -0.9 0.3]
```

$A$の固有値をみてみると実部に0.5799という正の値が含まれているため、開ループ系は不安定である。

```julia
eigvals(A)
```

```julia
B = [-0.2 0.1;
    -0.7 -0.6;
    0.6 0.9;
    -0.4 0.4]
```

```julia
C = [0.8 0.8 -0.8 -0.9;
    -0.7 0.8 -0.8 -0.2]
```

### 安定化を達成するゲイン行列の例

$K$を次のように取ると確かに$A + B K C$の固有値の実部はすべて負になる

```julia
K = [-2.7 2.7;
    2.7 -2.7]
```

```julia
eigvals(A + B * K * C)
```

確かに$A + B K C$の固有値の実部はたしかに全て負になっている

### グローバル変数の設定

```julia
T = 20.0
N = 500
η = T/N
```

### 常微分方程式を解く

微分方程式
$$
\frac{d}{dt} x(t)   = (A + B K C) x(t)
$$
を解くことでシステム応答を計算する

### オイラー法

```julia
function Euler(K)
    x = ones(4)
    x_trj = zeros(4, N)
    for i in 1:N
        x = x + η*(A + B*K*C)*x
        x_trj[:,i] = x
    end
    return x_trj[1,:], x_trj[2,:], x_trj[3,:], x_trj[4,:]
end        
```

### 解のプロット

```julia
x1,x2,x3,x4 = Euler(K);
plot(η:η:T, x1, label="x_1", xlabel="t",color=:black)
plot!(η:η:T, x2, label="x_2",color=:black, ls=:dot)
plot!(η:η:T, x3, label="x_3",color=:black, ls=:dash)
plot!(η:η:T, x4, label="x_4",color=:black, ls=:dashdot)
```

```julia
savefig("DU-stabilize-1.pdf")
```

### DU-Euler法

$\|x(T)\|^2$の値を損失関数として利用している。

```julia
function DU_Euler(K)
    x = ones(4)
    for i in 1:N
        x = x + η*(A + B*K*C)*x
    end
    return norm(x)^2
end        
```

### 学習プロセスと学習の結果 (学習の反復回数2回)

```julia
opt = ADAM(1e-0) 
train_itr = 2
K_train = ones(2,2)
ps = Flux.params(K_train)
for i in 1:train_itr
    gs = gradient(ps) do
        DU_Euler(K_train)
    end
    Flux.Optimise.update!(opt, ps, gs)
end
x1,x2,x3,x4 = Euler(K_train);
p1 = plot(η:η:T, x1, label="x_1", xlabel="t",color=:black)
plot!(η:η:T, x2, label="x_2",color=:black, ls=:dot)
plot!(η:η:T, x3, label="x_3",color=:black, ls=:dash)
plot!(η:η:T, x4, label="x_4",color=:black, ls=:dashdot)
```

```julia
savefig("DU-stabilize-2.pdf")
```

### 学習プロセスと学習の結果 (学習の反復回数20回)

```julia
opt = ADAM(1e-0) 
train_itr = 20
K_train = ones(2,2)
ps = Flux.params(K_train)
for i in 1:train_itr
    gs = gradient(ps) do
        DU_Euler(K_train)
    end
    Flux.Optimise.update!(opt, ps, gs)
end
x1,x2,x3,x4 = Euler(K_train);
p2 = plot(η:η:T, x1, label="x_1", xlabel="t",color=:black)
plot!(η:η:T, x2, label="x_2",color=:black, ls=:dot)
plot!(η:η:T, x3, label="x_3",color=:black, ls=:dash)
plot!(η:η:T, x4, label="x_4",color=:black, ls=:dashdot)
```
