```julia
using LinearAlgebra
using Plots
gr()
using Random
Random.seed!(1)
using Flux
```

### 準備 (グローバル定数の設定)

```julia
n = 128
m = 64
σ = 0.1
p = 0.05
num_itr = 40

K = 200
adam_lr = 5e-5
max_inner = 20

A = randn(m, n)
γ_opt = 1.0/maximum(eigvals(A'*A))
```

### データセット生成関数

```julia
function dataset(D)
    seq = randn(n,D)
    support = zeros(n, D)
    for i in 1:n
        for j in 1:D
            if rand() < p
                support[i,j] = 1
            end
        end
    end
    x = seq .* support
    y = A*x + σ*randn(m,D)
    return y, x
end
```

### ミニバッチ生成関数 (スパースベクトル生成関数)

無限データセット

```julia
function mini_batch(K)
    seq = randn(n,K)
    support = zeros(n, K)
    for i in 1:n
        for j in 1:K
            if rand() < p
                support[i,j] = 1
            end
        end
    end
    x = seq .* support
    y = A*x + σ*randn(m,K)
    return y, x
end
```

有限データセット

```julia
function mini_batch(D, K, dataset_y, dataset_x)
    x = zeros(n, K)
    y = zeros(m, K)
    for j in 1:K
        r = rand(1:D)
        x[:,j] = dataset_x[:,r]
        y[:,j] = dataset_y[:,r]
    end
    return y, x
end
```

```julia
function SoftTh(x, γ) 
    return softshrink(x, γ)
end
```

### ISTA法

```julia
function ISTA(max_itr, y)
    x = zeros(n, K)
    for i in 1:max_itr
        x = x - γ_opt*A'*(A*x - y)
        x = softshrink.(x, γ_opt)
    end
    return x
end

function comp_mse(max_itr)
    y, x_org = mini_batch(K)
    x̂ = ISTA(max_itr, y)
    return norm(x_org - x̂)^2/K
end
        
ISTA_mse = [comp_mse(i) for i in 1:num_itr]
plot(1:num_itr, ISTA_mse, lw=1, yscale=:log10, label="ISTA (fixed step size)", color=:black,  xlabel="Iteration", ylabel="Mean squared error",legend=:bottomleft)
```

### DU-ISTA反復の定義 

```julia
function DU_ISTA(max_itr, y, γ, α)
    x = zeros(n, K)
    for i in 1:max_itr
        x = x - γ[i]*A'*(A*x - y)
        x = softshrink.(x, α[i])
    end
    return x
end
```

### 損失関数の定義

```julia
function loss(x_org, x̂)
    return norm(x_org - x̂)^2/K
end     
```

### 訓練関数の定義

```julia
function train(opt, γ, α, dataset_y, dataset_x, K, D)
    ps = Flux.params(γ, α) 
    for itr in 1:num_itr 
        for i in 1:max_inner 
            y, x_org = mini_batch(D, K, dataset_y, dataset_x)
            gs = gradient(ps) do
                x_hat = DU_ISTA(itr, y, γ, α)
                loss(x_org, x_hat)
            end
            Flux.Optimise.update!(opt, ps, gs) 
        end
        y, x_org = mini_batch(K)
        x̂ = DU_ISTA(itr, y, γ, α)
        println(itr, " ", loss(x_org, x̂))
    end
end
```

### DU-GDのMSE評価用関数

```julia
function comp_mse_DUGD(max_itr, γ, α)
    y, x_org = mini_batch(K)
    x̂ = DU_ISTA(max_itr, y, γ, α)
    return loss(x_org, x̂)
end
```

### 学習過程

```julia
adam_lr = 1e-6
D = 10
dataset_y, dataset_x = dataset(D)
γ_train = γ_opt*ones(num_itr)
α_train = γ_opt*ones(num_itr)
opt = ADAM(adam_lr)
train(opt, γ_train, α_train, dataset_y, dataset_x, K, D)
DUGD_mse =[comp_mse_DUGD(i, γ_train, α_train) for i in 1:num_itr]
plot!(1:num_itr, DUGD_mse, lw=1, label="DU-ISTA D="*string(D), color=:black, marker=:+)
```

```julia
adam_lr = 5e-6
D = 50
dataset_y, dataset_x = dataset(D)
γ_train = γ_opt*ones(num_itr)
α_train = γ_opt*ones(num_itr)
opt = ADAM(adam_lr)
train(opt, γ_train, α_train, dataset_y, dataset_x, K, D)
DUGD_mse =[comp_mse_DUGD(i, γ_train, α_train) for i in 1:num_itr]
plot!(1:num_itr, DUGD_mse, lw=1, label="DU-ISTA D="*string(D), color=:black, marker=:star8)
```
