# MMSE推定関数のプロット

```julia
using LinearAlgebra
using Plots
gr()
```

```julia
F(x, var) =1/(sqrt(2pi*var))*exp(-(1/(2var))x^2)
function η(y, var, p) 
    numerator = p*F(y,1+var)
    denomitor = (1-p)*F(y, var) + p*F(y, 1+var)
    return (y/(1+var))*(numerator/denomitor)
end
```

```julia
xrange = -3:0.01:3
#plot(xrange, y->η(y, 1.0, 0.1), label="σ^2 = 1.0",legend=:topleft, xlabel="y")
plot(xrange, y->η(y, 0.5, 0.1), label="σ^2 = 0.5",legend=:topleft, xlabel="y", ls=:dash, c=:black)
plot!(xrange, y->η(y, 0.1, 0.1), label="σ^2 = 0.1", ls=:dot,c=:black)
plot!(xrange, y->η(y, 0.01, 0.1), label="σ^2 = 0.01", ls=:solid, c=:black)
```
